```python
# 1.  Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

import plotly.graph_objects as go
import plotly.express as px
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df.X, y = df.Y, z = df.Z, mode = 'markers'))
```

```python
# 4. using the given x and y data, generate a 3d surface plot to visualize the function z = sin((x**2 + y**2)1/2)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df.X, y = df.Y, z = df.Z))
fig.show()
```

Bokeh assignment


```python
# 1. create a Bokeh plot displaying a sine wave. set x valiues from 0 to 10 and y values as the sine of x
x = np.linspace(0, 10, 100)
y = np.sin(x)
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
p = figure(title = "sine wave")
p.line(x, y, line_width = 2)
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(5, 20, 100)
colors = np.random.choice(['red', 'green', 'blue'], 100)
p = figure(title = 'size_vs_colors')
p.circle(x, y, size = sizes, fill_color = colors)
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(title = 'bar chart', x_range = fruits, y_range = (0, 40))
p.vbar(x = fruits, top = counts, width = .2)
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title = 'histogram')
p.quad(top = hist, left = edges[:-1], right = edges[1:], bottom = 0, line_color = 'black')
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
p = figure(
    title = 'heatmap',
    x_axis_label = 'x axis',
    y_axis_label = 'y axis'
)
p.rect(data_heatmap, x =xx.flatten(), y = yy.flatten(), height = 0.1,width = 0.1)
show(p)
```
